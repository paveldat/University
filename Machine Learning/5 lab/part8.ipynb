```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('svmdata6.txt', delimiter='\t')
data.head()
```

```python
X = data.drop('Y', axis=1)
y = data.Y
```

```python
mse = []
eps_range = np.arange(0.01, 1.01, 0.01) 
for eps in eps_range:
    svr = SVR(epsilon=eps)
    svr.fit(X, y)
    mse.append(mean_squared_error(y, svr.predict(X)))
```

```python
plt.figure(figsize=(10,10))
plt.plot(eps_range, mse)
plt.xlabel('eps')
plt.ylabel('mse')
```

```python
svr = SVR(epsilon=0.01)
svr.fit(X, y)
plt.figure(figsize=(7,5))
plt.plot(X, y, label='Исходные данные')
plt.plot(X, svr.predict(X), label='Регрессионная кривая, epsilon = 0.01')
svr = SVR(epsilon=1)
svr.fit(X, y)
plt.plot(X, svr.predict(X), label='Регрессионная кривая, epsilon = 1')
plt.legend()
plt.tight_layout()
```
