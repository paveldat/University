```python
import numpy as np
import pandas as pd
import cv2 as cv
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
```

```python
def draw_results(data, n_clusters):
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    kmeans = KMeans(n_clusters=n_clusters).fit_predict(data)
    plt.scatter(data.col1, data.col2, c=kmeans, cmap='jet')
    plt.title('kmeans')
    plt.subplot(1, 3, 2)
    dbscan = DBSCAN(eps=0.4, min_samples=10).fit_predict(data)
    plt.scatter(data.col1, data.col2, c=dbscan, cmap='hsv')
    plt.title('DBSCAN')
    plt.subplot(1, 3, 3)
    aggl = AgglomerativeClustering(n_clusters=n_clusters).fit_predict(data)
    plt.scatter(data.col1, data.col2, c=aggl, cmap='seismic')
    plt.title('Hierarchical agglomerative')
    plt.tight_layout()
```

```python
data1 = pd.read_csv('clustering_1.csv', delimiter='\t', names=['col1', 'col2'])
_ = plt.scatter(data1.col1, data1.col2, c=np.ones(data1.col1.shape), cmap='flag')
```

```python
draw_results(data1, 2)
```

```python
data2 = pd.read_csv('clustering_2.csv', delimiter='\t', names=['col1', 'col2'])
_ = plt.scatter(data2.col1, data2.col2)
```

```python
draw_results(data2, 3)
```

```python
data3 = pd.read_csv('clustering_3.csv', delimiter='\t', names=['col1', 'col2'])
_ = plt.scatter(data3.col1, data3.col2)
```

```python
draw_results(data3, 2)
```
